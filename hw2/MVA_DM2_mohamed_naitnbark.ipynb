**(a) **   

- For a node $k \in E$ we denote $\pi_k$ the set of its parents in E and $\pi'_k$ the set of tis parents in $E'$.  
Let $p \in L(G)$, $p(x) = \prod_{k \in V}{p(x_k|x_{\pi_k})} = p(x_i|x_{\pi_i})p(x_j|x_{\pi_j})\prod_{k \in V-\{i,j\}}{p(x_k|x_{\pi_k})}$  
For $p \in L(G')$ $p(x) = \prod_{k \in V}{p(x_k|x_{\pi'_k})} = p(x_i|x_{\pi'_i})p(x_j|x_{\pi'_j})\prod_{k \in V-\{i,j\}}{p(x_k|x_{\pi_k})}$  Since $\forall k \in V-\{i,j\}: \pi_k = \pi'_k$     
$E' = (E-\{(i,j)\}) \cup \{(j,i)\}$ and $\pi_j = \pi_i \cup \{i\}$  implies that:  
$\pi'_j = \pi_i$  and  $\pi'_i = \pi'_j \cup \{j\} = \pi_i \cup \{j\}$   
To show that $L(G) = L(G')$  we need only to show that $p(x_i|x_{\pi_i})p(x_j|x_{\pi_j}) = p(x_i|x_{\pi'_i})p(x_j|x_{\pi'_j})$  
$$p(x_i|x_{\pi'_i}) = p(x_i|x_{\pi_i}, x_j) = \frac{p(x_j|x_{\pi_i},x_i)p(x_i|x_{\pi_i})}{p(x_j|x_{\pi_i})} = \frac{p(x_j|x_{\pi_i \cup \{i\}})p(x_i|x_{\pi_i})}{p(x_j|x_{\pi'_j})} = \frac{p(x_j|x_{\pi_j})p(x_i|x_{\pi_i})}{p(x_j|x_{\pi'_j})}$$  
Therefore $p(x_i|x_{\pi_i})p(x_j|x_{\pi_j}) = p(x_i|x_{\pi'_i})p(x_j|x_{\pi'_j})$  and  $L(G) = L(G')$

- $G = (V,E)$ a directed tree without v-structures, then $\forall i \in V, \pi_i$ is a singleton     
Let C be a click of $G'$, suppose that C contain more than three elements, then we have one of the two situations:  
i-  C contain a v-structure, this can't be in our case.  
ii- C doesn't contain any v-structures, in this case we have necessary a cycle which contradicts the fact that G is a tree.   
Therefore all the cliques contain exactly two elements and are of the form $ \{i\} \cup \pi_i$   
Let p be an element of $L(G')$:  
$p(x) = \frac{1}{Z} \prod_{i \in V}\psi_{i}(x_i,x_{\pi_i})$ such that  $\forall i \in V: \psi_i \geq 0$  and  $Z = \sum_{x} \psi_i$
Therefore $p \in L(G)$ and $L(G) = L(G')$

#### (a) K-mans algorithm:

```python
# Initial random centers
K = 4
centers = (0.5-rand(K,2))*array([8,10])
```

```python
def kmeans(train,centers):
    j=0
    K = shape(centers)[0]
    new_centers = zeros_like(centers)
    d = zeros((shape(train)[0],K))
    distortions = []
    while True:    
        for k in range(0,K):
            d[:,k] = norm(train - centers[k,:], axis = 1)
        points_centers = argmin(d,axis=1)
        distortions.append((d**2).min(axis=1).sum())
        for k in range(0,K):
            new_centers[k,:] = mean(train[points_centers == k], axis=0)
        if (centers == new_centers).all():
            return (centers, points_centers, array(distortions))
        else:
            centers = copy(new_centers)
```

```python
new_centers,points_centers,distortions = kmeans(train, centers)
```

The figure above represents the distortion at each step of the KMeans algorithm

#### (b) EM algorithm for a gaussian mixture with covariance matrices proportional to identity:

```python
from scipy.stats import multivariate_normal
from scipy.stats import norm as normal
from scipy.stats import chi2
```

```python
def em_gaussianmixture1(data, K, max_iter=1000):
    # Initialization:
    means = (0.5-rand(K,2))*array([8,10])
    means = kmeans(data,means)[0]
    variances = ones((K,2))
    p = repeat(1/float(K),K)
    resp = zeros((K,data.shape[0]))
    i=0
    # EM loop
    while i < max_iter:
        i = i + 1
        for k in range(0,K):
            resp[k,:] = p[k]*multivariate_normal.pdf(data, mean=means[k,:],cov=diag(variances[k,:]))
        resp = resp/resp.sum(axis=0)
        p = resp.mean(axis=1)
        resp = (resp.T/resp.sum(axis=1)).T
        means = resp.dot(data)
        for k in range(0,K):
            variances[k,:] = resp[k,:].dot((data - means[k,:])**2)
    return (p, means, variances)
```

```python
def gaussian_mixture_classify(data, p, means, covs):
    N = data.shape[0]
    d = p.shape[0]
    probs = zeros((N,d))
    for k in range(0,d):
        probs[:,k] = p[k]*multivariate_normal.pdf(data, mean=means[k,:],cov=covs[k])
    return argmax(probs, axis=1)
```

#### (c) EM algorithm for a gaussian mixture in the general case:

```python
def em_gaussianmixture2(data, K, max_iter=1000):
    # Initialization:
    means = (0.5-rand(K,2))*array([8,10])
    means = kmeans(data,means)[0]
    cov = [diag([1,1]) for i in range(0,4)]
    p = repeat(1/float(K),K)
    resp = zeros((K,data.shape[0]))
    i=0
    # EM loop
    while i < max_iter:
        i = i + 1
        for k in range(0,K):
            resp[k,:] = p[k]*multivariate_normal.pdf(data, mean=means[k,:],cov=cov[k])
        resp = resp/resp.sum(axis=0)
        p = resp.mean(axis=1)
        resp = (resp.T/resp.sum(axis=1)).T
        means = resp.dot(data)
        for k in range(0,K):
            cov[k] = (resp[k,:]*((data - means[k,:]).T)).dot(data - means[k,:])
    return (p, means, cov)
```

The general gaussian mixture model appears to perform well on the test data, we can confirm that by calculating the loglikelihood

```python
def gaussian_mixture_loglikelihood(data,means, cov, p):
    K = p.size
    dens = zeros((data.shape[0], K))
    for k in range(0,K):
        dens[:,k] = multivariate_normal.pdf(data, mean=means[k,:], cov=cov[k])
    dens = dens.dot(p)
    return log(dens).sum()
```
