## I - Learning in discrete graphical models

**Maximum likelihood for the r.a. z:**   
Let's code our $ (z_m)_{m \in{[1,M]}} $ by vectors of dimension $ M $ such that $ \forall m, i \in [1,M] : z_m^{(i)} = \delta_{m,i}$ where $\delta$ is the Kronecker symbol and $ z_m^{(i)} $ the ith components of the vector $z_m$.   
Then we can write $P(z) = \prod_{m \in [1,M]}{\pi_m^{z^{(m)}}}$ and then our likelihood function is: $L(z_1,..,z_n,\pi) = \prod_{i \in [1,N]}{\prod_{m \in [1,M]}{\pi_m^{z_i^{(m)}}}}$.
Hence the log-likelihood is $l(z_1,..,z_n,\pi) = \sum_{i \in [1,N]}{\sum_{m \in [1,M]}{z_i^{(m)}log(\pi_m)}}$.
Then we can formulate the maximum likelihood estimator $\pi^{MV}$ as the solution of the problem:$$argmax_{\pi}{\sum_{i \in [1,N]}{\sum_{m \in [1,M]}{z_i^{(m)}log(\pi_m)}}} \\s.t.   \forall m \in [1,M]: \pi_m \geq 0\   and   \sum_{m \in [1,M]}{\pi_m} = 1$$
$$l(z_1,..,z_n,\pi) = \sum_{i \in [1,N]}{\sum_{m \in [1,M]}{z_i^{(m)}log(\pi_m)}} = \sum_{m \in [1,M]}{(\sum_{i \in [1,n]}{z_i^{(m)})log(\pi_m)}} = \sum_{m \in [1,M]}{n_{m}log(\pi_m)}$$ Where $n_m = card\{i | z_i^{m} = 1\}$     
Then the lagrangian of our problem is $L(\pi,\lambda) = - \sum_{m \in [1,M]}{n_{m}log(\pi_m)} + \lambda (\sum_{k \in [1,K]}{\pi_k} - 1)$      
$L(\pi,\lambda)$ is a convex function, and for example $\frac{1}{K}$ **1** $\in ]0;+\infty[^K \cap \{\pi | sum_{k \in [1,K]}{pi_k} = 1\}$ then by Slater's constraint qualification lemma:     
$\max_{\pi}{l(\pi)} = \max_{\lambda}{\min_{\pi}{L(\pi,\lambda)}}$       
$\frac{\partial L}{\partial \pi_{m}} = - \frac{\pi_m}{n_m} + \lambda \Rightarrow \pi_m = \frac{n_m}{\lambda}$      
The constraint $\sum_{m \in [1,M]}{\pi_m} = 1$ gives $\lambda = n$ therefore: $$\forall m \in [1,M]: \pi_m = \frac{n_m}{n}$$

**Maximum likelihood for the r.a. x:**   
An analogous derivation as below shows that our log-likelihood is given by: $l(x_1,..,x_n|z_1,...,z_n,\theta) = \sum_{i \in [1,n]}{\sum_{m \in [1,M]}{\sum_{k \in [1,K]}{x_i^{(k)}z_i^{(m)}log(\theta_m^k)}}}$.
Again we can write the loglikelihood in the form: $l(x_1,..,x_n|z_1,...,z_n,\theta) = \sum_{k \in [1,K]}{\sum_{m \in [1,M]}{(\sum_{i \in [1,n]}{x_i^{(k)}z_i^{(m)}})log(\theta_m^k)}} = \sum_{k \in [1,K]}{\sum_{m \in [1,M]}{n_m^k log(\theta_m^k)}}$.      
Where $n_m^k = \sum_{i \in [1,n]}{x_i^{(k)}z_i^{(m)}}$.      
This is the same problem as above and by using the same arguments as in the case of $z$ we derive the MV estimator ($n_m = \sum_{k \in [1,K]}{n_m^k}$):$$\theta_m^k = \frac{n_m^k}{n_m}$$

## II - Linear classification

### 1 - Generative model (LDA)

** (a) **       
The probability of having a realisation $(x,y)$ is $p(x,y;\pi;\mu) = p(x|y;\pi;\mu) = \prod_{k \in [1,K]}{(\pi_k f{(x,\mu_k)})^{y^{(k)}}}$ where $f{(x,\mu_k)}$ is the normal density function $N(\mu_k, \Sigma)$.       
Then the log-likelihood of our model is given by:
$$l(\pi,\mu) = \sum_{n \in [1,N]}{\sum_{k \in [1,K]}{y_n^{(k)}[-\frac{1}{2}(x_n - \mu_k)^T \Sigma^{-1} (x_n - \mu_k) + log(\pi_k)]}} - Nlog(2\pi) - \frac{N}{2}log(\det{\Sigma}) \\ s.t \sum_{k \in [1,K]}{\pi_k} = 1$$
The lagrangian of this problem is $L(\pi,\mu,\lambda) = - \sum_{n \in [1,N]}{\sum_{k \in [1,K]}{y_n^{(k)}[-\frac{1}{2}(x_n - \mu_k)^T \Sigma^{-1} (x_n - \mu_k) + log(\pi_k)]}} + \lambda (\sum_{k \in [1,K]}{\pi_k} - 1)$.    
The lagrangian is clearly a convex function and the condition of the Slater's constraint lemma is obviously verified, so we can write:
$$\max_{\pi,\mu}l(\pi,\mu) = \max_{\lambda}{\min_{\pi,\mu}{L(\pi,\mu,\lambda)}}$$
$\forall k \in [1,K]: \frac{\partial L}{\partial \pi_k} = \frac{\sum_{n \in [1,N]}{y_n^{(k)}}}{2 \pi_k} + \lambda = 0 \Rightarrow \pi_k = \frac{\sum_{n \in [1,N]}{y_n^{(k)}}}{2 \lambda}$     
Since $\sum_{k \in [1,K]}{\pi_k} = 1$ we get $\lambda = \frac{N}{2}$ and: $$\forall k \in [1,K]: \pi_k = \frac{n_k}{N}\\  n_k = \sum_{n \in [1,N]}{y_n^{(k)}}$$
$\forall k \in [1,K]: \frac{\partial L}{\partial \mu_k} = \sum_{n \in [1,N]}{y_n^{(k)}(x_n - \mu_k)} = 0 \Rightarrow \mu_k = \frac{\sum_{n \in [1,N]}{y_n^{(k)} x_n}}{N}$

** (b) **      
In the case of y Bernoulli r.a. :       
Using Bayes formula: $$P(y=1|x) = \frac{P(x|y=1)P(y=1)}{P(x|y=1)P(y=1) + P(x|y=2)P(y=2)}$$
i.e. $$P(y=1|x) = \frac{\pi_1\exp{-(\frac{(x - \mu_1)^T \Sigma^{-1} (x - \mu_1)}{2})}}{\pi_1\exp{-(\frac{(x - \mu_1)^T \Sigma^{-1} (x - \mu_1)}{2})} + \pi_2\exp{-(\frac{(x - \mu_2)^T \Sigma^{-1} (x - \mu_2)}{2})}} = \frac{\frac{\pi_1}{\pi_2}\exp{(\beta x + \alpha)}}{\frac{\pi_1}{\pi_2}\exp{(\beta x + \alpha)} + 1}$$
Where $\beta = \Sigma^{-1}(\mu_1 - \mu_2)$ and $\alpha = \frac{1}{2}(\mu_2\Sigma^{-1}\mu_2 - \mu_1\Sigma^{-1}\mu_1)$.     
This model corresponds to a logistic regression with parameters $\beta' = \beta$ and $\alpha' = \alpha + ln(\frac{\pi_1}{\pi_2})$

### 2 - Logistic regression

```python
#Load data
from scipy.special import expit
tab_train = loadtxt('./classification_data_HWK1/classificationA.train', delimiter='\t')
```

```python
def split_data(data):
    X = ones((data.shape[0],data.shape[1]))
    X[:,:-1] = data[:,(0,1)]
    Y = data[:,2]
    return X,Y
```

```python
# Separate the features from the target variables 
X,Y = split_data(tab_train)
```

```python
#Implement the Newton Ralphson algorithm
def logreg_train(X,Y):
    #Init
    beta_logreg = (0,0,0)
    threshold = 0.001
    likelihood = []
    #Loop
    while True:
        likelihood.append(Y.dot(log(expit(X.dot(beta_logreg)))) +(1-Y).dot(log(1-expit(X.dot(beta_logreg)))))
        beta_prev = beta_logreg
        J = X.T.dot(Y - expit(X.dot(beta_logreg)))
        R = diag( expit(X.dot(beta_logreg))*(1 - expit(X.dot(beta_logreg))) )
        H = -(X.T.dot(R)).dot(X)
        beta_logreg = beta_logreg - solve(H,J)
        if norm(beta_logreg - beta_prev)/norm(beta_prev) < threshold:
            break
    return beta_logreg, likelihood
```

```python
beta_logreg, likelihood = logreg_train(X,Y)
```

```python
plot(likelihood)
```

The graph above represents the log likelihood in each iteration of the optimization algorithm

### 5 - QDA model

Using bayes formula we derive the expression: $$P(y=1|x) = \frac{1}{\exp{-(x^T\textbf{A}x + \textbf{B}x+\alpha)}+1}$$        
Where $\textbf{A} = \frac{1}{2}(\Sigma_2^{-1} - \Sigma_1^{-1})$ and $\textbf{B} = \mu_1^T\Sigma_1^{-1} - \mu_2^T\Sigma_2^{-1}$ and $\alpha = \frac{1}{2}(\mu_2^T\Sigma_2^{-1}\mu_2 - \mu_1^T\Sigma_1^{-1}\mu_1)$

** (a) **

```python
# This function trains a QDA model
def qda_train(X,Y):
    class1 = X[Y == 1]
    class2 = X[Y == 0]
    p1 = mean(Y==1)
    mean1 = class1.mean(0)
    mean2 = class2.mean(0)
    sigma1 = (class1-mean1).T.dot(class1-mean1)/len(class1)
    sigma2 = (class2-mean2).T.dot(class2-mean2)/len(class2)
    sigma1_inv = inv(sigma1)
    sigma2_inv = inv(sigma2)
    A = 0.5*(sigma2_inv - sigma1_inv)
    B = mean1.T.dot(sigma1_inv) - mean2.T.dot(sigma2_inv)
    alpha = 0.5*(mean2.T.dot(sigma2_inv).dot(mean2) - mean1.T.dot(sigma1_inv).dot(mean1)) + log(p1/(1-p1))
    return A,B,alpha
```

```python
# Train QDA model on the trainC dataset
A,B,alpha = qda_train(X_c[:,(0,1)],Y_c)
```
